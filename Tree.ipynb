```python
import pandas as pd
```

```python
df=pd.read_csv("titanic-passengers.csv",sep=";")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df["Cabin"]=df.Cabin.fillna("G6")
```

```python
df["Embarked"]=df.Embarked.fillna("S")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
df["Survived"]=encoder.fit_transform(df["Survived"])
df
```

```python
encoder=LabelEncoder()
df["Sex"]=encoder.fit_transform(df["Sex"])
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
x=df.drop(["Survived","Cabin","Ticket","Embarked","Name",],axis=1)
y=df["Survived"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=10)
myTree=tree.DecisionTreeClassifier()
myTree = myTree.fit(x_train,y_train)
y_pred=myTree.predict(x_test)
print("score:{}".format(accuracy_score(y_test,y_pred)))
```

```python
import graphviz
dot_df = tree.export_graphviz(myTree, out_file=None)
graph = graphviz.Source(dot_df)
graph.render("df")
graph
```

```python
x=df.drop(["Survived","Cabin","Ticket","Embarked","Name","PassengerId","SibSp","Fare"],axis=1)
y=df["Survived"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=20)
myTree=tree.DecisionTreeClassifier()
myTree = myTree.fit(x_train,y_train)
y_pred=myTree.predict(x_test)
print("score:{}".format(accuracy_score(y_test,y_pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics 
```

```python
clf=RandomForestClassifier(n_estimators=10)
clf = clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
